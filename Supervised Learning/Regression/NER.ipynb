```python
def filter1(result, stopwords, min_length=4, min_score=0.80):
    filtered=[]
    for i in result:
        word=i["word"]
        score=i["score"]

        if word.startswith("##"):
            continue
        if len(word)<min_length:
            continue
        if score<min_score:
            continue
        if word.lower() in stopwords:
            continue

        filtered.append({"word": word, "score": float(score)})

    return filtered
```
