```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['brand','model','vehicle_age','km_driven','seller_type','fuel_type','transmission_type','mileage','engine','max_power','seats']]
Y=df['selling_price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
OH=OneHotEncoder(drop='first',handle_unknown='infrequent_if_exist')
ST=StandardScaler()
categorical=['model','brand','seller_type','fuel_type','transmission_type']
numeric=['vehicle_age','km_driven','mileage','engine','max_power','seats']
```

```python
pipe=ColumnTransformer([("OneHotEncoder",OH,categorical),("StandardScaler",ST,numeric)])
```

```python
X_train=pipe.fit_transform(X_train)
X_test=pipe.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg=RandomForestRegressor(n_estimators=100,max_depth=10,random_state=42)
```

```python
reg.fit(X_train,Y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
```

```python
r2=r2_score(Y_test,y_pred,)
mse=mean_squared_error(Y_test,y_pred)
mae=mean_absolute_error(Y_test,y_pred)
print(r2)
print(mse)
print(mae)
print(np.sqrt(mse))
```

```python
trees=[100,200,300,400,500]
depths=[5,10,15,20]
for tree in trees:
    for depth in depths:
        reg=RandomForestRegressor(n_estimators=tree,max_depth=depth,random_state=42)
        reg.fit(X_train,Y_train)
        y_pred=reg.predict(X_test)
        r2=r2_score(Y_test,y_pred)
        mse=mean_squared_error(Y_test,y_pred)
        mae=mean_absolute_error(Y_test,y_pred)
        print(r2)
        print(mse)
        print(mae)
        print(np.sqrt(mse))
        print()
```

```python
param={
    "n_estimators":[100,200,300,400,500],
    "max_depth":[10,15,20,25]
}
```

```python
Ran=RandomizedSearchCV(reg,param_distributions=param,cv=5,n_jobs=-1)
Ran.fit(X_train,Y_train)
```

```python
best_model = Ran.best_estimator_
final_pred = best_model.predict(X_test)
print(r2_score(Y_test, final_pred))
```
