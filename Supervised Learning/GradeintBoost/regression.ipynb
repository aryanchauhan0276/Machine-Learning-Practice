```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['brand','model','vehicle_age','km_driven','seller_type','fuel_type','transmission_type','mileage','engine','max_power','seats']]
Y=df['selling_price']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
OH=OneHotEncoder(drop='first',handle_unknown='infrequent_if_exist')
ST=StandardScaler()
categorical=['model','brand','seller_type','fuel_type','transmission_type']
numeric=['vehicle_age','km_driven','mileage','engine','max_power','seats']
```

```python
pipe=ColumnTransformer([("OneHotEncoder",OH,categorical),("StandardScaler",ST,numeric)])
```

```python
X_train=pipe.fit_transform(X_train)
X_test=pipe.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
Reg=GradientBoostingRegressor(n_estimators=100,max_depth=15)
```

```python
Reg.fit(X_train,Y_train)
```

```python
pred=Reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2=r2_score(Y_test,pred)
mse=mean_squared_error(Y_test,pred)
mae=mean_absolute_error(Y_test,pred)
print(r2)
print(mse)
print(mae)
print(np.sqrt(mse))
```

```python
ada={
    "n_estimators":[30,50,80,100],
    "loss":['huber', 'quantile', 'squared_error', 'absolute_error']
}
```
