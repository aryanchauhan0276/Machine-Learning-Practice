```python
from sklearn.model_selection import train_test_split
```

```python
X=df[['Age','TypeofContact','CityTier','DurationOfPitch','Occupation','Gender','NumberOfFollowups','ProductPitched','PreferredPropertyStar','MaritalStatus','NumberOfTrips','Passport','PitchSatisfactionScore','OwnCar','Designation','MonthlyIncome','Member Visiting']]
Y=df['ProdTaken']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
numeric=['Age','CityTier','DurationOfPitch','NumberOfFollowups','PreferredPropertyStar','NumberOfTrips','Passport','PitchSatisfactionScore','OwnCar','MonthlyIncome','Member Visiting']
categorical=['TypeofContact','Occupation','Gender','ProductPitched','MaritalStatus','Designation']
```

```python
numeric_transformer = StandardScaler()
oh_transformer = OneHotEncoder(drop='first')
preprocessor=ColumnTransformer([
    ("OneHotEncoder",oh_transformer,categorical),("StandardScaler",numeric_transformer,numeric)
])
```

```python
X_train=preprocessor.fit_transform(X_train)
```

```python
X_test=preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
Grad=GradientBoostingClassifier(n_estimators=100)
```

```python
Grad.fit(X_train,Y_train)
```

```python
pred=Grad.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
acc=accuracy_score(Y_test,pred)
clas=classification_report(Y_test,pred)
print(acc)
print(clas)
print(confusion_matrix(Y_test,pred))
```
