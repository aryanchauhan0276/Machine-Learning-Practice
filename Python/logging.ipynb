```python
import logging
import os
import time
```

```python
logging.basicConfig(filename="app.log",filemode='w',level=logging.DEBUG)
```

```python
logging.debug("This is debugging statement")
logging.info("This is information statement")
logging.warning("This is warning statement")
logging.error("This is error message")
logging.critical("This is a critical message")
```

```python

def add(a,b):
    size=os.path.getsize("app.log")
    for handler in logging.root.handlers[:]:
        logging.root.removeHandler(handler)
   
    result=a+b
    print(result)
    print(size)
    if size > 400:
        logging.basicConfig(filename="new.log",filemode='a',level=logging.DEBUG)
        logging.debug("This is being written into new file")
    else:
            logging.basicConfig(filename="app.log",filemode='a',level=logging.DEBUG)
            logging.debug("Additon has been done in old file")

        

add(10,20)

def sub(a,b):
    size=os.path.getsize("app.log")
    for handler in logging.root.handlers[:]:
        logging.root.removeHandler(handler)
    subtract=abs(a-b)
    print(subtract)
    print(size)
    if size > 400:
        logging.basicConfig(filename="new.log",filemode='a',level=logging.DEBUG)

        logging.debug("This is being written into new file")
    else:
        logging.basicConfig(filename="app.log",filemode='a',level=logging.DEBUG)
            
        logging.debug("Substraction has been done into old file")
sub(23,45)
```
